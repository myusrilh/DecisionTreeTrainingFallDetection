```python
# source : https://www.meccanismocomplesso.org/en/3d-rotations-and-euler-angles-in-python/
import numpy as np
import math as m

# Roll
def Rz(theta):
    return np.matrix([[ 1, 0           , 0           ],
                      [ 0, m.cos(theta),-m.sin(theta)],
                      [ 0, m.sin(theta), m.cos(theta)]])

# Pitch
def Rx(theta):
    return np.matrix([[ m.cos(theta), 0, m.sin(theta)],
                      [ 0           , 1, 0           ],
                      [-m.sin(theta), 0, m.cos(theta)]])

# Yaw
def Ry(theta):
    return np.matrix([[ m.cos(theta), -m.sin(theta), 0 ],
                      [ m.sin(theta), m.cos(theta) , 0 ],
                      [ 0           , 0            , 1 ]])
```

```python
# source: https://developpaper.com/the-difference-between-atan-and-atan2-in-math-module-of-python/

# Atan and atan2 are arctangent functions that return radians

# For the straight line formed by two points, the two points are point (x1, Y1) and point (X2, Y2), and the calculation method of slope corresponding angle can be as follows:

# angle = atan( (y2-y1)/(x2-x1) )
# or
# angle = atan2( y2-y1, x2-x1 )

# Therefore, we can see the difference between atan and atan2

# 1. The number of parameters is different; atan is a single parameter, atan2 is two parameters
# 2. The advantage of atan2 is that if x2-x1 is equal to 0, the angle can still be calculated, 
#    but atan function needs to be judged in advance, otherwise it will lead to program error;

# Conclusion: atan and atan2 functions should be used;

# Note: 1. Both return values are radians; 2. Atan2 function parameters are y before X; 3. You need to import math library first
# Let’s look at their usage and differences according to several examples:

# import math
# a=math.atan(1)
# b=math.atan2(1,1)
# print(a,b)
# >> 0.7853981633974483 0.7853981633974483

# #Vertical
# b=math.atan2(1,0)
# print(b)
# >> 1.5707963267948966

# #Radian rotation angle
# b=math.atan2(1,0)
# print(b/math.pi*180)
# >> 90.0

# The above detailed explanation of the differences between atan and atan2 under math module in Python is the whole content shared by Xiaobian. 
# I hope it can give you a reference and support developeppaer more
        
```

```python
# Source : https://github.com/rocheparadox/Kalman-Filter-Python-for-mpu6050/blob/master/Kalman.py
class KalmanAngle:
    def __init__(self):
        self.QAngle = 0.001
        self.QBias = 0.003
        self.RMeasure = 0.03
        self.angle = 0.0
        self.bias = 0.0
        self.rate = 0.0
        self.P=[[0.0,0.0],[0.0,0.0]]

    '''def kalman():
        QAngle = 0.001
        QBias = 0.003
        RMeasure = 0.03
        angle = 0.0
        bias = 0.0
        P[0][0] = 0.0
        P[0][1] = 0.0
        P[1][0] = 0.0
        P[1][1] = 0.0'''

    def getAngle(self,newAngle, newRate,dt):
        #step 1:
        self.rate = newRate - self.bias;    #new_rate is the latest Gyro measurement
        self.angle += dt * self.rate;

        #Step 2:
        self.P[0][0] += dt * (dt*self.P[1][1] -self.P[0][1] - self.P[1][0] + self.QAngle)
        self.P[0][1] -= dt * self.P[1][1]
        self.P[1][0] -= dt * self.P[1][1]
        self.P[1][1] += self.QBias * dt

        #Step 3: Innovation
        y = newAngle - self.angle

        #Step 4: Innovation covariance
        s = self.P[0][0] + self.RMeasure

        #Step 5:    Kalman Gain
        K=[0.0,0.0]
        K[0] = self.P[0][0]/s
        K[1] = self.P[1][0]/s

        #Step 6: Update the Angle
        self.angle += K[0] * y
        self.bias  += K[1] * y

        #Step 7: Calculate estimation error covariance - Update the error covariance
        P00Temp = self.P[0][0]
        P01Temp = self.P[0][1]

        self.P[0][0] -= K[0] * P00Temp;
        self.P[0][1] -= K[0] * P01Temp;
        self.P[1][0] -= K[1] * P00Temp;
        self.P[1][1] -= K[1] * P01Temp;

        return self.angle

    def setAngle(self,angle):
        self.angle = angle

    def setQAngle(self,QAngle):
        self.QAngle = QAngle

    def setQBias(self,QBias):
        self.QBias = QBias

    def setRMeasure(self,RMeasure):
        self.RMeasure = RMeasure

    def getRate():
        return self.rate

    def getQAngle():
        return self.QAngle

    def getQBias():
        return self.QBias

    def  getRMeasure():
        return self.RMeasure
```

### First version (Kalman Filter)

```python
# Arah sensor : 
# 1) x positif ke kiri (pitch), y positif ke atas (yaw), z positif ke arah lawan pengguna (roll)
# 2) x positif ke kiri (pitch), y positif ke bawah (yaw), z positif ke arah pengguna (roll)
# source : https://www.meccanismocomplesso.org/en/3d-rotations-and-euler-angles-in-python/

# Sample Frequency = 238 Hz (238 data per second)
# Total time per file = 20 seconds
# Delta t = 1/f = 1/238
# Use only last 13 seconds per file is enough (it's already includes pre-fall, free-fall, impact shock, body adjustment and post-fall phases)

# Delete data with D1 and D2 on filename

# Convert accel to (g - m/s^2) and gyro to (degree per second)
# Acc_Sen= 0.000244; # Accelerometer sensitivity 0.244 mg/LSB
# Gyr_Sen= 0.07;     # Angular rate sensitivity 70 mdps/LSB
# Mag_Sen= 0.00014;  # Magnetic sensitivity 0.14 mgauss/LSB

# data bernilai negatif jika arahnya searah gravitasi

# from math import atan2
from numpy import pi
# import time
import math

directory = os.getcwd()+"\\..\\SelectedData\\CleanData\\"
new_directory = os.getcwd()+"\\..\\SelectedData\\CleanDataAgain\\"

raw_files = os.listdir(directory)

n_freq = 238
d_time = 1/n_freq
rad_to_deg = 180/pi
comp_const = 1/(1+d_time)

kalmanZ = KalmanAngle()
kalmanX = KalmanAngle()

kalAngleZ = 0
kalAngleX = 0

for f in raw_files:
    if 'D1' in f or 'D2' in f:
        os.remove(os.path.join(directory,f)) #delete file with D1 and D2 name (Neck and Wrist device)
        print(f, "deleted!")
    else:
        
        acc_sen= 0.000244 # hasil dari -> 8/32767.5
        gyr_sen= 0.07 # hasil dari -> 2300/32767.5
        
        dataframe = pd.read_csv(os.path.join(directory,f), header=None, names=["Ax","Ay","Az","Gx","Gy","Gz","C1","C2","Acc_roll","Acc_pitch","Gyro_roll","Gyro_pitch","Roll","Pitch","Yaw"])
        
        
        for ind in range(len(dataframe)):
            dataframe.loc[ind,"Ax"] = dataframe.loc[ind,"Ax"]*acc_sen
            dataframe.loc[ind,"Ay"] = dataframe.loc[ind,"Ay"]*acc_sen
            dataframe.loc[ind,"Az"] = dataframe.loc[ind,"Az"]*acc_sen
            
            # C1 = Sum vector magnitude on horizontal plane [x,z] (np.sqrt(np.power(Ax ke-i,2) + np.power(Az ke-i,2)))
            dataframe.loc[ind,"C1"] = np.sqrt(np.power(dataframe.loc[ind,"Ax"],2) + np.power(dataframe.loc[ind,"Az"],2))

            # C2 = Sum vector magnitude [x,y,z] (np.sqrt(np.power(Ax ke-i,2) + np.power(Ay ke-i,2) + np.power(Az ke-i,2)))
            dataframe.loc[ind,"C2"] = np.sqrt(np.power(dataframe.loc[ind,"Ax"],2) + np.power(dataframe.loc[ind,"Ay"],2) + np.power(dataframe.loc[ind,"Az"],2))

            dataframe.loc[ind,"Acc_roll"] = math.atan2(dataframe.loc[ind,"Az"],dataframe.loc[ind,"Ay"])*rad_to_deg        
            dataframe.loc[ind,"Acc_pitch"] = math.atan(-(dataframe.loc[ind,"Ax"])/(np.sqrt(np.power(dataframe.loc[ind,"Az"],2) + np.power(dataframe.loc[ind,"Ay"],2))))*rad_to_deg
            
            dataframe.loc[ind,"Yaw"] = dataframe.loc[ind,"Yaw"]+(dataframe.loc[ind,"Gy"]*d_time)
            if dataframe.loc[ind,"Yaw"] < 0:
                dataframe.loc[ind,"Yaw"] = dataframe.loc[ind,"Yaw"]+360.0
                    
            roll = dataframe.loc[ind,"Acc_roll"]
            pitch = dataframe.loc[ind,"Acc_pitch"]
            
            kalmanZ.setAngle(roll)
            kalmanX.setAngle(pitch)
            
            gyroZAngle = roll
            gyroXAngle = pitch
            
            gyroZRate = dataframe.loc[ind,"Gz"]*gyr_sen
            gyroXRate = dataframe.loc[ind,"Gx"]*gyr_sen
            
            if((roll < -90 and kalAngleX >90) or (roll > 90 and kalAngleX < -90)):
                kalmanZ.setAngle(roll)
                complAngleZ = roll
                kalAngleX   = roll
                gyroZAngle  = roll
            else:
                kalAngleZ = kalmanZ.getAngle(roll,gyroZRate,d_time)

            if(abs(kalAngleZ)>90):
                gyroXRate  = -gyroXRate
                kalAngleX  = kalmanX.getAngle(pitch,gyroXRate,d_time)
            
            gyroZAngle = gyroZRate * d_time
            gyroXAngle = gyroXRate * d_time

            if ((gyroZAngle < -180) or (gyroZAngle > 180)):
                gyroZAngle = kalAngleZ
            if ((gyroXAngle < -180) or (gyroXAngle > 180)):
                gyroXAngle = kalAngleX
            
            dataframe.loc[ind,"Gyro_roll"] = gyroZAngle
            dataframe.loc[ind,"Gyro_pitch"] = gyroXAngle
            dataframe.loc[ind,"Roll"] = kalAngleZ
            dataframe.loc[ind,"Pitch"] = kalAngleX
        
        
        dataframe.to_csv(os.path.join(new_directory,f), sep=",", header=False, index=False)
        print(f,' saved!')
```

```python
# Confusion Matrix
# https://medium.com/analytics-vidhya/confusion-matrix-accuracy-precision-recall-f1-score-ade299cf63cd

from sklearn.metrics import confusion_matrix, accuracy_score, recall_score, precision_score

# confusion matrix
confusion = confusion_matrix(y_test, y_pred)
# print(confusion)

TP = confusion[0][0]
FP = confusion[0][1]
FN = confusion[1][0]
TN = confusion[1][1]

# Method 1: Sklearn
# accuracy score
# akurasi = accuracy_score(y_test, y_pred)

# recall score
# recall = recall_score(y_test, y_pred, average='micro')

# precision score
# presisi = precision_score(y_test, y_pred, average='micro')

# Method 2: Manual

# Confusion matrix:
# [[TP,FP],
#  [FN,TN]] -> TP = True Positive, FP = False Positive, FN = False Negative, TN = True Negative

# Accuracy/Akurasi
akurasi = (TN + TP) / (TP+FP+FN+TN)

# Precision/Presisi -> Menghitung positive predictive value
presisi = TP/(TP+FP)

# Recall -> Recall is also known as sensitivity or true positive rate
recall = TP/(TP+FN)

error_rate = (FN + FP) / (TP+FP+FN+TN)

print("Akurasi: ",akurasi)
print("Presisi: ",presisi)
print("Recall: ",recall)
print("Error Rate: ",error_rate)

confusion
```
